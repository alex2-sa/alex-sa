```python
#comparer les isocodes de notre excel et les excel du bti et récupérer la valeur lorsque les 2 isocodes se correspondent
bti_2018 = bti_df[bti_df['year'] == 2018]
for index, isocode in df.isocode.iteritems(): #df.isocode est une autre maière d'appeler la code isocode de df plutot que [""]
    try:
        ind = bti_2018[bti_2018['country_code'] == isocode]['stat_ind'] #on parcoure les 2 colonnes d'isocode et on récupère la valeur de l'indicateur
        rank = bti_2018[bti_2018['country_code'] == isocode]['rank_stat_ind'] 
        df.at[index,'Bertelsmann Transformation index (BTI)'] = ind #on met le résultat trouvé dans la bonne colonne du Excel
        df.at[index,'Bertelsmann Transformation index (BTI) - Ranking'] = rank
    except ValueError: 
        print('[critical]:no data for',isocode) 
        continue
df[['Country (according to CN Gov):','Bertelsmann Transformation index (BTI)','Bertelsmann Transformation index (BTI) - Ranking']]
                            
```

```python
heritage_df = pd.read_excel('C:\\Users\\Alex\\Documents\\UTC\\UTSEUS\\US01\\data-heritage-2019.xls')
```

```python
heritage_df.head()
```

```python
for index, name in df['Country (according to CN Gov):'].iteritems():
    cleaned_name = name.split('(')[0].strip()
    if cleaned_name == 'Thailiand':
        cleaned_name = 'Thailand'
    elif cleaned_name == 'The United Arab Emirates':
        cleaned_name = 'United Arab Emirates'
    elif cleaned_name == 'Brunei':
        cleaned_name = 'Brunei Darussalam'
    elif cleaned_name == 'Myanmar':
        cleaned_name = 'Burma'
    elif cleaned_name == 'East Timor':
        cleaned_name = 'Timor-Leste'
    elif cleaned_name == 'Kyrgyzstan':
        cleaned_name = 'Kyrgyz Republic'
    elif cleaned_name == 'United States of America':
        cleaned_name = 'United States'
    
    
    try:
        ind = heritage_df[heritage_df['Country Name']==cleaned_name]['2019 Score']
        rank = heritage_df[heritage_df['Country Name']==cleaned_name]['World Rank']
        df.at[index,'HF Index of Economic Freedom (Scale: 1-100)'] = ind
        df.at[index,'Heritage Foundation Index of Economic Freedom (Global Ranking)'] = rank
    except ValueError:
        print('[critical]:no data for',cleaned_name)
        continue
```

```python
ratings = pd.read_csv('C:\\Users\\Alex\\Documents\\UTC\\UTSEUS\\US01\\data-ratings-2015.csv', sep =';') #cette librairie lit fichier texte séparé par ',' et certains csv sont séparés par des points virgules on doit donc spécifier cela que le fichier utilisé 
ratings_2015 = ratings[ratings['Year'] == 2015]
ratings_2015[['Country2Code', 'SP_NumericRating','Moodys_NumericRating','Fitch_NumericRating']]
```

```python
for index, isocode in df.isocode.iteritems():
    try:
        sp = ratings_2015.loc[ratings_2015['Country2Code'] == isocode]['SP_NumericRating']
        moody = ratings_2015.loc[ratings_2015['Country2Code'] == isocode]['Moodys_NumericRating']
        fitch = ratings_2015.loc[ratings_2015['Country2Code'] == isocode]['Fitch_NumericRating']
        df.at[index,'S&P Sovereign Ratings List'] = sp
        df.at[index,"Moody's Sovereign Ratings List"] = moody
        df.at[index,'Fitch Sovereign Ratings List'] = fitch
    except ValueError:
        print('[critical]: no data for', isocode)
        continue
df[['S&P Sovereign Ratings List', "Moody's Sovereign Ratings List", 'Fitch Sovereign Ratings List']]
```
